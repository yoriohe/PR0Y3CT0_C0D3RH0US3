```python
import pandas as pd
import json 
import os
import requests
```

```python
## Configuración de la API de NewsAPI
api_key = 'e33f69379d5e4313b5cfa4ca52441644'  
url = 'https://newsapi.org/v2/top-headlines'
params = {'country': 'ar', 'apiKey': api_key}
```

```python
# Genero el get para verificar la respuesta 
df = requests.get(url,params=params)
df
```

```python
# Verifico el string obtendido
texto = df.text
texto
```

```python
# Ahora obtengo el objeto json  y convierto de dataframe
jsondata = json.loads(texto)
df2 = pd.DataFrame.from_dict(jsondata)
df2.head()
```

```python
# Crear un DataFrame a partir de la información de los repositorios
repositorios = []
for repo in df2['articles']:
        repositorios.append({
            'Autor': repo['author'],
            'Titulo': repo['title'],
            'URL': repo['url'],
            
           
        })
    
# Convertir la lista de diccionarios a un DataFrame
df_repos = pd.DataFrame(repositorios)
 
```
