## Feature Engeenering

```python
# Como siempre Guardo base original para cualquier consulta que quiero hacer
df1 = df.copy(deep = True)
df1.head()
```

```python
#Elimino previamente la columna customerID
columns_to_drop = ['customerID']
df1 = df1.drop(columns=columns_to_drop, errors='ignore')
```

```python
#Buscamos aquellos features que no son numéricos si no se encuentran en la función describe.
text_data_features = [i for i in list(df1.columns) if i not in list(df1.describe().columns)]
text_data_features
```

```python
le = LabelEncoder()
for i in text_data_features :
    df1[i] = le.fit_transform(df1[i])
    print(i,' : ',df1[i].unique(),' = ',le.inverse_transform(df1[i].unique()))
```

```python
df1.describe()
```

```python
#Busco ver cuantos servicios de internet tiene cada cliente. Creo columna.

df1["Cant_Internet"] = df1["InternetService"] + df1["OnlineSecurity"] + df1["OnlineBackup"] + df1["DeviceProtection"] + df1["TechSupport"] + df1["StreamingTV"] + df1["StreamingMovies"]

# Cambiar cualquier valor igual a 'No internet service' a 0
df1["Cant_Internet"] = df1["Cant_Internet"].apply(lambda x: 0 if x == 'No internet service' else x)

# Cambiar cualquier valor mayor a 1 a 1
df1["Cant_Internet"] = df1["Cant_Internet"].clip(upper=1)
```

```python
# Creo la columna para saber si tienen servicio telefónico
df1["PhoneServices"]= df1["PhoneService"]+df1["MultipleLines"]
#Creo columna para ver quienes tienen internet + telefono.
df1["pack"]= df1["PhoneServices"] + df1["InternetService"]
```

```python
df1.info()
```

```python
# Elimina las columnas no numéricas antes de calcular la correlación
df_numeric = df1.select_dtypes(include=['float64', 'int64'])
df_numeric = df1.select_dtypes(include=['int', 'int64'])
```

```python
df_numeric.info()
```

```python
# Calcula la matriz de correlación
matriz_correlacion = df_numeric.corr(method='pearson')
matriz_correlacion
```

```python
# Selecciona la variable target para estudiar únicamente correlación con ella 
matriz_correlacion_target = matriz_correlacion['Churn'].abs()
```

```python
# Excluye columnas específicas del análisis de correlación
columnas_a_excluir = ['Churn']
matriz_correlacion_target = matriz_correlacion_target.drop(columns=columnas_a_excluir, index=columnas_a_excluir)
matriz_correlacion_target
```

```python
# Ordena de mayor a menor correlación 
matriz_correlacion_target = matriz_correlacion_target.sort_values(ascending=False)
# Muestra la matriz de correlación
print(matriz_correlacion_target)
```

```python
#Selección basada en correlación:
# Establecer un umbral de correlación
threshold = 0.1  # Ajusta el umbral según tus necesidades

# Seleccionar características con correlación por encima del umbral
selected_features = matriz_correlacion_target[abs(matriz_correlacion_target) > threshold].index

# Imprimir o utilizar las características seleccionadas
print(selected_features)
```

```python
# Seleccionar solo las características relevantes en df1
df1_selected = df1[selected_features]

# Añadir la columna de la variable objetivo ('Churn') a df1_selected si es necesario
df1_selected['Churn'] = df1['Churn']

# Verificar el nuevo DataFrame df1_selected
print(df1_selected.head())
```

```python
#Separamos en train y test
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(df1_selected,
                                    test_size = 0.2, #Porcentaje de muestra test 
                                    random_state = 43) #Semilla
df_train = df_train.reset_index()
df_test = df_test.reset_index()
```

```python
# Normalizamos los datos
from sklearn.preprocessing import StandardScaler
X_train = df_train[selected_features].values
X_test = df_test[selected_features].values
# Defino escalador
sc = StandardScaler()
# Genero parámetros con train
X_train = sc.fit_transform(X_train)
# Le aplico el escalador a test
X_test = sc.transform(X_test)
```

```python
print(f" DF ENTRENAMIENTO -->{df_train.shape}  \n DF TEST --> {df_test.shape}")
```

```python
print(df1_selected.columns)
```

```python
# Seleccionar las características con correlación significativa
features_to_include = ['Contract', 'tenure', 'OnlineSecurity', 'TechSupport', 'OnlineBackup',
       'PaperlessBilling', 'DeviceProtection', 'Dependents', 'SeniorCitizen',
       'Partner', 'PaymentMethod', 'Churn']

# Filtrar el conjunto de datos original utilizando solo las características seleccionadas
X_selected = df1[features_to_include]

# Aplicar PCA en el conjunto de datos seleccionado
from sklearn.decomposition import PCA

# Iniciar el objeto PCA
pca = PCA()

# Generar componentes con el conjunto de datos seleccionado
PCA_data = pca.fit_transform(X_selected)

# Almacenar en un DataFrame y presentar
PCA_df = pd.DataFrame(data=PCA_data, columns=[f'PC_{i+1}' for i in range(PCA_data.shape[1])])

# Mostrar el DataFrame resultante
PCA_df.head()
```

```python
# Análisis de la varianza explicada para cada componente
varianza_explicada = pca.explained_variance_ratio_
componentes = PCA_df.columns

plt.bar(componentes, 
        varianza_explicada)
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.title('Varianza Explicada por Componente Principal')
ax = plt.gca()
for p in ax.patches:
    x_text = p.get_x() + p.get_width() / 2.0  # Calcula la posición x del texto
    y_text = p.get_height()  # Toma la altura de la columna
    ax.text(x_text, y_text, f'{str(round(y_text * 100,2))}%', # Indica posición x e y, y texto a mostrar 
            ha = 'center', # Alineación horizontal 
            va = 'bottom',  # Alineación vertical
            fontsize = 10, color = 'black', fontweight = 'bold')
plt.show()
```
